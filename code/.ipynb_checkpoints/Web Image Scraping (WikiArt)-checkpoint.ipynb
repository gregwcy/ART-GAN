```python
# Create a list to house the URLs
url_list = [url1, url2, url3, url4, url5, url6, url7, url8, url9, url10,
           url11, url12, url13, url14, url15, url16, url17, url18, url19, url20,
           url21, url22, url23, url24, url25, url26, url27, url28, url29, url30,]
```

```python
# Code for scraping

# Counter to keep track of nmber of works scraped per artist
counter_list = []

# Function to scrape
def download_images(url_list):
    for v,i in enumerate(url_list):
        artist_name = artists[v]
        driver.get(i)
        
        # Set the driver to click load more to get full list of works
        while True:
            try:
                loadMoreButton = driver.find_element_by_xpath("//*[@ng-click='loadMore($event);']")
                time.sleep(5)
                loadMoreButton.click()
                time.sleep(15)
            except Exception as e:
                print(e)
                break
        print("Complete")
        time.sleep(10)
        
        # Find elements that contain the hyerlinks from thumbnails
        imgresults = driver.find_elements_by_xpath('//img[contains(@class, "ng-isolate-scope")]')
        totalresults = len(imgresults)
        
        # Get the source links
        src_links = [imgresults[i].get_attribute('src') for i in range(len(imgresults))]
        
        rows = []
        counter = 0
        
        # Download images from each link in source links
        for i in src_links:
            file_name = artist_name + str(counter) + '.jpg'
            print(file_name)
            response = requests.get(i)
            file = open(file_name, "wb")
            file.write(response.content)
            file.close()
            counter += 1
        counter_list.append(counter)
        
```

```python
# Run function on the list of urls
download_images(url_list)
```

```python
# Create Dataframe to house info on number of works scraped per artist
works = pd.DataFrame(columns = ['Artist', 'Num_Works'])
works['Artist'] = artists
works['Num_Works'] = counter_list
new_row = {'Artist': 'Total', 'Num_Works': works['Num_Works'].sum()}
works = works.append(new_row, ignore_index=True)
works
```
