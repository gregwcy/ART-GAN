In data augmentation, from our existing samples, we generate new samples via transformation from our original sample.
This helps us to address potential overfitting to the original sample, we can also train our machine on a larger dataset.


```python
import os
from PIL import Image
import numpy as np
import imgaug as ia
from imgaug import augmenters as iaa
import imageio
import matplotlib.pyplot as plt
```

```python
# We run a sample on augmenting
image = Image.open('../data/raw_images/Mark Rothko1.jpg')
print(image.size) # width x height 
print(image.format)
print(image.mode)
plt.imshow(image);
```

There are a few things to note here. We see that the image size is not uniform, we will have to resize all images.
We next move on to perform several transformations. This will allow us to produce 6 images from 1.
1. The original
2. Rotated 90 degrees
3. Rotated 180 degrees
4. Rotated 270 degrees
5. Flipped left to right
6. Flipped top to bottom

```python
# We first resize the image

image = image.resize((64,64), resample=Image.BICUBIC)  # We use bicubic here as it is deemed to produce better results

plt.figure(figsize=(10,10))
plt.subplot(3,2,1)
plt.title('Resized')
plt.imshow(image)
plt.subplot(3,2,2)
plt.title('Flipped left right')
plt.imshow(image.transpose(Image.FLIP_LEFT_RIGHT))
plt.subplot(3,2,3)
plt.title('Flipped top bottom')
plt.imshow(image.transpose(Image.FLIP_TOP_BOTTOM))
plt.subplot(3,2,4)
plt.title('Rotated 90 degrees')
plt.imshow(image.transpose(Image.ROTATE_90))
plt.subplot(3,2,5)
plt.title('Rotated 180 degrees')
plt.imshow(image.transpose(Image.ROTATE_180))
plt.subplot(3,2,6)
plt.title('Rotated 270 degrees')
plt.imshow(image.transpose(Image.ROTATE_270))
```

```python
# We run a tiral with 3 images and test out our function

# Set up directories
source = '../data/sample'
destination = '../data/sample_augmented'
path, dirs, files = next(os.walk(source))

# Set up params
size = 64 

# Set up image augmenter

for file in sorted(files):
    print(file)
    image = Image.open(path + '/' + file)
    if image.mode == "RGB":
        image = image.resize((size,size), resample=Image.BICUBIC)
        to_file = destination + "/" + file[:-4] + '.jpg'
        image.save(to_file)
        
        # Flip Left Right
        image_lr = image.transpose(Image.FLIP_LEFT_RIGHT)
        to_file = destination + "/" + file[:-4] + ' LR' + '.jpg'
        image_lr.save(to_file)
        
        # Flip Top Bottom
        image_tb = image.transpose(Image.FLIP_TOP_BOTTOM)
        to_file = destination + "/" + file[:-4] + ' TB' + '.jpg'
        image_tb.save(to_file)
        
        # Rotate 90 degrees
        image_90 = image.transpose(Image.ROTATE_90)
        to_file = destination + "/" + file[:-4] + ' 90' + '.jpg'
        image_90.save(to_file)
        
        # Rotate 180 degrees
        image_180 = image.transpose(Image.ROTATE_180)
        to_file = destination + "/" + file[:-4] + ' 180' + '.jpg'
        image_180.save(to_file)
        
        # Rotate 270 degrees
        image_270 = image.transpose(Image.ROTATE_270)
        to_file = destination + "/" + file[:-4] + ' 270' + '.jpg'
        image_270.save(to_file)
    
```

Our code works and now we can apply it on the entire dataset

```python
# Set up directories
source = '../data/raw_images'
destination = '../data/images_augmented'
path, dirs, files = next(os.walk(source))

# Set up params
size = 1024 

# Set up image augmenter

for file in sorted(files):
    print(file)
    image = Image.open(path + '/' + file)
    if image.mode == "RGB":
        image = image.resize((size,size), resample=Image.BICUBIC)
        to_file = destination + "/" + file[:-4] + '.jpg'
        image.save(to_file)
        
        # Flip Left Right
        image_lr = image.transpose(Image.FLIP_LEFT_RIGHT)
        to_file = destination + "/" + file[:-4] + ' LR' + '.jpg'
        image_lr.save(to_file)
        
        # Flip Top Bottom
        image_tb = image.transpose(Image.FLIP_TOP_BOTTOM)
        to_file = destination + "/" + file[:-4] + ' TB' + '.jpg'
        image_tb.save(to_file)
        
        # Rotate 90 degrees
        image_90 = image.transpose(Image.ROTATE_90)
        to_file = destination + "/" + file[:-4] + ' 90' + '.jpg'
        image_90.save(to_file)
        
        # Rotate 180 degrees
        image_180 = image.transpose(Image.ROTATE_180)
        to_file = destination + "/" + file[:-4] + ' 180' + '.jpg'
        image_180.save(to_file)
        
        # Rotate 270 degrees
        image_270 = image.transpose(Image.ROTATE_270)
        to_file = destination + "/" + file[:-4] + ' 270' + '.jpg'
        image_270.save(to_file)
    
```

We originally had 937 images but after augmenting, we now have 5622 (937 x 6). 

After sifting through the new dataset, some images could not be used, due to the original odd sizing, resulting in just white boxes. We cleaned those and now are left with a total of 5,562 images.
